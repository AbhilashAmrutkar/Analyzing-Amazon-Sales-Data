```python
df1 = pd.read_csv('C:\ABHILASH\Data Science\Project\Project 2\Amazon Sales Records R1.csv')
```

```python
df1.drop(labels = 'Unnamed: 0', axis=1, inplace= True)
```

```python
df1
```

```python
# Sales by year
sales_by_Order_Date = df1.groupby("Order Date")["Total Revenue"].sum()
```

```python
sales_by_Order_Date
```

```python
# plot the data
plt.plot(sales_by_Order_Date)
plt.xlabel("Order Date")
plt.ylabel("Total Revenue")
plt.title("Sales Trend Over Time")
plt.show()
```

```python
# convert the "Order date" column to a Pandas datetime object
df1["Order Date"] = pd.to_datetime(df1["Order Date"])
```

```python
# extract the day, month, and year from the "order date" column
df1["o_day"] = df1["Order Date"].dt.day
df1["o_month"] = df1["Order Date"].dt.month
df1["o_year"] = df1["Order Date"].dt.year
```

```python
df1
```

```python
# Sales by month
sales_by_month = df1.groupby("o_month")["Total Revenue"].sum()
```

```python
# plot the data
plt.plot(sales_by_month)
plt.xlabel("o_month")
plt.ylabel("Total Revenue")
plt.title("Sales Trend Over Time")
plt.show()
```

```python
# Sales by year
sales_by_month = df1.groupby("o_year")["Total Revenue"].sum()
```

```python
# plot the data
plt.plot(sales_by_month)
plt.xlabel("o_year")
plt.ylabel("Total Revenue")
plt.title("Sales Trend Over Time")
plt.show()
```

```python
# group the data by year and month and sum the sales for each year/month combination
sales_by_year_month = df1.groupby(["o_year", "o_month"])["Total Revenue"].sum()
```
